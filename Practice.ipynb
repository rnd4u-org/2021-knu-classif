Import pandas library and find out info about data frame from "train.csv"

```python
import pandas
train=pandas.read_csv("train.csv")
print("size: ", train.size, "\n")
print("shape: ", train.shape, "\n")
print("columns: ", list(train.columns), "\n")
print("type of each column:\n", train.dtypes)
```

Print out our data frame

```python
train
```

Remove all NaNs

```python
import numpy as np

train = train.dropna()
train
```

Import matplotlib and for every categorical variable build a barplot

```python
import matplotlib.pyplot as plt

columns = train.columns[-3::]

fig, axs = plt.subplots(1, 3, figsize=(20,  5))

for i in range(3):
    value_dict = dict(train[columns[i]].value_counts(sort=False))
    axs[i].set_title(f"{columns[i]} barplot")
    axs[i].bar(value_dict.keys(), value_dict.values())



plt.show()
```

For every numerical value build boxplot

```python
col_names = list(train.columns[1:-3])

plt.boxplot(train[col_names])

plt.show()
```

Plot histograms

```python
fig, axs = plt.subplots(2, 4, figsize=(20,  10))

col_ind = 0

for i in range(2):
    for j in range(4):
        axs[i, j].set_title(f"{col_names[col_ind]} histogram")
        axs[i, j].hist(train[col_names[col_ind]])
        col_ind += 1

plt.show()
```

Normallize data 

```python
normalized_train = train.copy()
normalized_train[col_names] = normalized_train[col_names].apply(lambda x: (x-x.min())/(x.max()-x.min()))
normalized_train
```

Print out boxplot after normalization

```python

fig, axs = plt.subplots()

axs.boxplot(normalized_train[col_names])
axs.set_xticklabels(col_names, rotation=90)
plt.show()
```

Print histograms after data normalization

```python
fig, axs = plt.subplots(2, 4, figsize=(20,  20))

col_ind = 0

for i in range(2):
    for j in range(4):
        axs[i, j].set_title(f"{col_names[col_ind]} histogram")
        axs[i, j].hist(normalized_train[col_names[col_ind]])
        col_ind += 1
plt.show()
```

Encode categorical features so that i can use it in ML and drop first column as it doesnt tell us anything

```python
from sklearn.preprocessing import LabelEncoder

categorical_cols = normalized_train.columns[-3::]
le = LabelEncoder()
normalized_train[categorical_cols] = normalized_train[categorical_cols].apply(lambda col: le.fit_transform(col))
main_train = normalized_train[normalized_train.columns[1::]]
main_train
```

# Random Forest

Removing outliers

```python
from scipy import stats
ml_train = main_train.copy()
num_train = ml_train[ml_train.columns[:-3]]
num_train = num_train[(np.abs(stats.zscore(num_train)) < 3).all(axis=1) == True]
indexes = list(num_train.index)

ml_train = ml_train.loc[indexes]
ml_train[ml_train.columns[:-3]] = num_train

ml_train
```

```python
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X = ml_train[ml_train.columns[:-1]]
y = ml_train['customer_category']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

regressor = RandomForestClassifier()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

# Feature Selection

```python
from sklearn.feature_selection import SelectFromModel

feat_sel = SelectFromModel(RandomForestClassifier())
feat_sel.fit(X_train, y_train)

selected_feat= X_train[X_train.columns[(feat_sel.get_support())]]
reg2 = RandomForestClassifier()
reg2.fit(X_train, y_train)

y2_pred = reg2.predict(X_test)
print(feat_sel.get_support())
print("New Accuracy: ", metrics.accuracy_score(y_test, y2_pred))
```
