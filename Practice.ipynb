# Random Forest

Removing outliers

```python
from scipy import stats
ml_train = main_train.copy()
num_train = ml_train[ml_train.columns[:-3]]
num_train = num_train[(np.abs(stats.zscore(num_train)) < 3).all(axis=1) == True]
indexes = list(num_train.index)

ml_train = ml_train.loc[indexes]
ml_train[ml_train.columns[:-3]] = num_train

ml_train
```

```python
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

X = ml_train[ml_train.columns[:-1]]
Y = ml_train['customer_category']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)

regressor = RandomForestClassifier()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

# Feature Selection

```python
from sklearn.feature_selection import SelectFromModel

feat_sel = SelectFromModel(RandomForestClassifier())
feat_sel.fit(X_train, y_train)

selected_feat= X_train[X_train.columns[(feat_sel.get_support())]]
reg2 = RandomForestClassifier()
reg2.fit(X_train, y_train)

y2_pred = reg2.predict(X_test)
print(feat_sel.get_support())
print("New Accuracy: ", metrics.accuracy_score(y_test, y2_pred))
```

# Model Evaluation, ROC Curve

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

randforest = RandomForestClassifier()
nearneighbor = KNeighborsClassifier(n_neighbors=6)
naivebayes = GaussianNB()
supvector = SVC(kernel="linear", probability=True)
logregression = LogisticRegression()

models = [
    ["Random Forest ", randforest],
    ["K neighbors ", nearneighbor],
    ["Naive Bayes ", naivebayes], 
    ["Support vector machine ", supvector],
    ["Regression ", logregression]
]

pred_values = []
proba_values = []
roc_auc_score_values = []
roc_curve_values = []
cross_val_scores = []

for model in models:
    print(models.index(model) * 100 // len(models), "% ...")
    model[1].fit(X_train, y_train)
    
    pred = model[1].predict(X_test)
    proba = model[1].predict_proba(X_test)[:, 1]
    auc = metrics.roc_auc_score(y_test, proba)
    pr, tpr, thresholds = metrics.roc_curve(y_test, proba)
    
    cross_score = cross_val_score(model[1], X, Y, cv=10, scoring="accuracy").mean()
    
    pred_values.append([model[0], pred])
    proba_values.append([model[0], proba])
    roc_auc_score_values.append([model[0], auc])
    roc_curve_values.append([model[0], [pr, tpr]])
    cross_val_scores.append([model[0], cross_score])
    
print("Training done")
```

```python
print("Accuracy for each algorithm:\n")
for value in pred_values:
    print(value[0], metrics.accuracy_score(y_test, value[1]))

print("\n")
    
print("Area under each curve after each algorithm:\n")
for value in roc_auc_score_values:
    print(*value)
    
print("\n")
    
print("Cross validation accuracy for each algorithm:\n")
for value in cross_val_scores:
    print(*value)
    
for value in roc_curve_values:
    plt.plot(value[1][0], value[1][1], marker=".", label=value[0])
    
plt.xlabel("False Positive")
plt.ylabel("True Positive")
plt.legend()
plt.show()
```
