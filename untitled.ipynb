```python
from scipy import stats
dt=dt[['customer_visit_score', 'customer_product_search_score', 'customer_ctr_score', 'customer_stay_score', 'customer_frequency_score', 'customer_product_variation_score', 'customer_order_score', 'customer_affinity_score','customer_active_segment','X1','customer_category']]
dt = dt[(np.abs(stats.zscore(dt)) < 3).all(axis=1)]
dt
```

```python
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
X = dt[['customer_visit_score', 'customer_product_search_score', 'customer_ctr_score', 'customer_stay_score', 'customer_frequency_score', 'customer_product_variation_score', 'customer_order_score', 'customer_affinity_score','customer_active_segment','X1']]
y = dt['customer_category']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
regressor = RandomForestClassifier(n_estimators=200,min_samples_split=4,min_samples_leaf=3,max_features=7,max_depth=10)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
nbor = KNeighborsClassifier(n_neighbors=6)
nbor.fit(X_train, y_train)
nbayes = GaussianNB()
nbayes.fit(X_train, y_train)
svector = SVC(kernel="linear", probability=True)
svector.fit(X_train, y_train)
logregression = LogisticRegression()
logregression.fit(X_train, y_train)
y_nbor_pred = nbor.predict(X_test)
y_nbayes_pred = nbayes.predict(X_test)
y_svector_pred = svector.predict(X_test)
y_logregression_pred = logregression.predict(X_test)
print("Accuracy RF:",metrics.accuracy_score(y_test, y_pred))
print("Precision RF:",metrics.precision_score(y_test, y_pred))
print("Recall RF:",metrics.recall_score(y_test, y_pred))
print("______")
print("Accuracy Nbor:",metrics.accuracy_score(y_test, y_nbor_pred))
print("Accuracy Nbayes:",metrics.accuracy_score(y_test, y_nbayes_pred))
print("Accuracy Svector:",metrics.accuracy_score(y_test, y_svector_pred))
print("Accuracy Logregression:",metrics.accuracy_score(y_test, y_logregression_pred))
```

```python
#Feature_Selection
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import Ridge
ridge = Ridge().fit(X, y)
model = SelectFromModel(ridge, prefit=True)
X_transformed = model.transform(X)
X_transformed[:5] # its customer_ctr_score and customer_stay_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
regressor2 = RandomForestClassifier(n_estimators=400,min_samples_split=3,min_samples_leaf=2,max_features=8,max_depth=10)
regressor2.fit(X_train, y_train)
y_new_pred = regressor2.predict(X_test)
print("Accuracy RF:",metrics.accuracy_score(y_test, y_new_pred))
```

```python
#ROC
y_pred_proba = regressor.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="RF, auc="+str(auc))
plt.legend(loc=4)
y_nbor_proba = nbor.predict_proba(X_test)[::,1]
fprnbor, tprnbor, _ = metrics.roc_curve(y_test, y_nbor_proba)
aucnbor = metrics.roc_auc_score(y_test, y_nbor_proba)
plt.plot(fprnbor,tprnbor,label="Nbor, auc="+str(aucnbor))
plt.legend(loc=4)
y_nbayes_proba = nbayes.predict_proba(X_test)[::,1]
fprnbayes, tprnbayes, _ = metrics.roc_curve(y_test, y_nbayes_proba)
auÑnbayes = metrics.roc_auc_score(y_test, y_nbayes_proba)
plt.plot(fprnbayes,tprnbayes,label="Nbayes, auc="+str(auÑnbayes))
plt.legend(loc=4)
y_svector_proba = svector.predict_proba(X_test)[::,1]
fprsvector, tprsvector, _ = metrics.roc_curve(y_test, y_svector_proba)
aucsvector = metrics.roc_auc_score(y_test, y_svector_proba)
plt.plot(fprsvector,tprsvector,label="Svector, auc="+str(aucsvector))
plt.legend(loc=4)
y_logregression_proba = logregression.predict_proba(X_test)[::,1]
fprlogregression, tprlogregression, _ = metrics.roc_curve(y_test, y_logregression_proba)
auclogregression = metrics.roc_auc_score(y_test, y_logregression_proba)
plt.plot(fprlogregression,tprlogregression,label="Logregression, auc="+str(auclogregression))
plt.legend(loc=4)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
print('CVS RF',np.mean(cross_val_score(regressor, X, y)))
print('CVS Nbor',np.mean(cross_val_score(nbor, X, y,)))
print('CVS Nbayes',np.mean(cross_val_score(nbayes, X, y)))
print('CVS Svector',np.mean(cross_val_score(svector, X, y)))
print('CVS Logregression',np.mean(cross_val_score(logregression, X, y)))
```
