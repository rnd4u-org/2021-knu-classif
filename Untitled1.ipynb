```python
import pandas

ds_train = pandas.read_csv("train.csv")
ds_train = pandas.DataFrame(ds_train)   

print("There are {} rows in ds_train.".format(ds_train.shape[0]))
print("There are {} columns in ds_train.\n".format(ds_train.shape[1]))

i = 0
for col in ds_train.columns:
    print("Column {}:".format(i))
    print("Name: {} Type: {}".format('"' + col + '",', ds_train.dtypes[i]))
    i += 1

print("Size: {}".format(ds_train.size))

ds_train
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,3,figsize=(10,  5))
a=0
for i in ds_train.columns[9:12]:
    bar_dict = dict(ds_train[i].value_counts())
    ax[a].bar(bar_dict.keys(),bar_dict.values())
    ax[a].set_title(i)
    a+=1
plt.show()
```

```python

for col in ds_train.columns: 
    ds_train[col] = ds_train[col].fillna(ds_train.mean())
    
ncol = ds_train.columns[1:-3]
ds_train[ncol] = ds_train[ncol].apply(lambda x: (x-x.min())/(x.max()-x.min()))
ds_train
```
